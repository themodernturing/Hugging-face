### Build the `text-to-speech` pipeline using the ðŸ¤— Transformers Library

- Here is some code that suppresses warning messages.

````python
from transformers.utils import logging

logging.set_verbosity_error()
````

````python
# Load model directly
from transformers import AutoTokenizer, AutoModelForTextToWaveform

tokenizer = AutoTokenizer.from_pretrained("kakao-enterprise/vits-ljs")
model = AutoModelForTextToWaveform.from_pretrained("kakao-enterprise/vits-ljs")
````

````python
from transformers import pipeline

narrator = pipeline("text-to-speech",
                    model="kakao-enterprise/vits-ljs")
````

Info about [kakao-enterprise/vits-ljs](https://huggingface.co/kakao-enterprise/vits-ljs)

````python
text = """
Scientists at MIT, Stanford, Google, and Harvard have \
collaborated to create a new AI system that improves \
the efficiency of solar energy collection. The system uses \
advanced algorithms to optimize the placement and \
angle of solar panels based on weather patterns and \
geographical data. The study highlights the importance \
of accurate predictions in maximizing energy output, with \
panel positioning and timing being crucial factors.
"""
````

````python
narrated_text = narrator(text)
````

````python
from IPython.display import Audio as IPythonAudio

IPythonAudio(narrated_text["audio"][0],
             rate=narrated_text["sampling_rate"])
````
